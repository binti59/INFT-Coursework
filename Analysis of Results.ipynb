```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("GaussianFunction.csv")
```

```python
df['Kval']
```

```python

# Setup matplotlib
import matplotlib.pyplot as plt

# Create an axes object
axes = plt.gca()
x_ticks = [2, 3, 4, 5]
# pass the axes object to plot function
df.plot(kind='line', x='Kval', y='S1=0', ax=axes,color='r',marker='D');
df.plot(kind='line', x='Kval', y='S2=0.75', ax=axes,marker='D');
plt.title('Gaussian Function')
plt.xticks(ticks=x_ticks)
plt.xlabel('K-Value')
plt.ylabel('Average Profitability/Trader')
plt.grid()
plt.show()
```
